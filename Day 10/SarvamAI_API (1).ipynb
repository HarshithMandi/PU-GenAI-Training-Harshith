```python
from sarvamai import SarvamAI

client = SarvamAI(
    api_subscription_key="",
)

response = client.chat.completions(
    model="sarvam-105b",
    messages=[
        {"role": "user", "content": "Explain the economic impact of GST implementation in India."}
    ],
    temperature=0.5,
    top_p=1,
    max_tokens=4000,
)

print(response.choices[0].message.content)
```
